```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
drift_fields = [33.3, 34.0, 35.3, 35.3, 36.6]  # in kV/cm
timing_resolutions = [33.6, 29.1, 25.7, 27.2, 22.9]  # in ps
errors = [0.2, 0.1, 0.3, 0.3, 0.2]

# Field configurations (drift/amplification in kV/cm)
field_configs = ["33.3/22.6", "34.0/22.6", "35.3/22.6", "35.3/21.4", "36.6/21.4"]

# Plot
plt.figure(figsize=(10, 6))
for i, config in enumerate(field_configs):
    plt.errorbar(drift_fields[i], timing_resolutions[i], yerr=errors[i],
                 fmt='o', label=f'Config {config}', capsize=5)

plt.xlabel("Drift Field [kV/cm]")
plt.ylabel("Timing Resolution [ps]")
plt.title("Timing Resolution vs Drift Field")
plt.grid(True)
plt.legend(title="Field Config (Drift/Ampl.)")
plt.tight_layout()
plt.show()
```

```python
# Raw data
drift_fields = [33.3, 34.0, 35.3, 35.3, 36.6]
ampl_fields = [22.6, 22.6, 22.6, 21.4, 21.4]
timing_resolutions = [33.6, 29.1, 25.7, 27.2, 22.9]
errors = [0.2, 0.1, 0.3, 0.3, 0.2]

# Organize data by unique amplification field
unique_ampl_fields = sorted(set(ampl_fields))
data_by_ampl = {a: {"x": [], "y": [], "err": []} for a in unique_ampl_fields}

for x, a, y, e in zip(drift_fields, ampl_fields, timing_resolutions, errors):
    data_by_ampl[a]["x"].append(x)
    data_by_ampl[a]["y"].append(y)
    data_by_ampl[a]["err"].append(e)

# Plot
plt.figure(figsize=(10, 6))
for a_field in unique_ampl_fields:
    d = data_by_ampl[a_field]
    plt.errorbar(d["x"], d["y"], yerr=d["err"], fmt='o-', capsize=8, label=f'Ampl. Field: {a_field} kV/cm')

plt.xlabel("Drift Field [kV/cm]")
plt.ylabel("Timing Resolution [ps]")
# plt.title("Timing Resolution vs Drift Field")
plt.grid(True)
plt.legend(title="Amplification Field")
plt.tight_layout()
plt.show()
```

```python
# Set ROOT-like style
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 14,
    "axes.linewidth": 1.2,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 6,
    "ytick.major.size": 6,
    "xtick.minor.size": 3,
    "ytick.minor.size": 3,
    "xtick.top": True,
    "ytick.right": True,
    "legend.frameon": True,
    "legend.edgecolor": "black"
})

# Data
drift_fields = [33.3, 34.0, 35.3, 35.3, 36.6]
ampl_fields = [22.6, 22.6, 22.6, 21.4, 21.4]
timing_resolutions = [33.6, 29.1, 25.7, 27.2, 22.9]
errors = [0.2, 0.1, 0.3, 0.3, 0.2]

# Organize data by unique amplification field
unique_ampl_fields = sorted(set(ampl_fields))
data_by_ampl = {a: {"x": [], "y": [], "err": []} for a in unique_ampl_fields}

for x, a, y, e in zip(drift_fields, ampl_fields, timing_resolutions, errors):
    data_by_ampl[a]["x"].append(x)
    data_by_ampl[a]["y"].append(y)
    data_by_ampl[a]["err"].append(e)

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
colors = ['black', 'red', 'blue', 'green']
markers = ['o', 's', '^', 'D']

for i, a_field in enumerate(unique_ampl_fields):
    d = data_by_ampl[a_field]
    ax.errorbar(d["x"], d["y"], yerr=d["err"], fmt=markers[i]+'-',
                capsize=5, label=f'Ampl. Field: {a_field} kV/cm',
                color=colors[i], markersize=6, linewidth=1.5)

ax.set_xlabel("Drift Field [kV/cm]")
ax.set_ylabel("Timing Resolution [ps]")
ax.set_title("Timing Resolution vs Drift Field", fontsize=16)
ax.grid(False)
ax.legend(title="Amplification Field", loc='upper right', fontsize=12)

# Tweak spines to match ROOT
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
