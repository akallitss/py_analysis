```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
drift_fields = [33.3, 34.0, 35.3, 35.3, 36.6]  # in kV/cm
timing_resolutions = [33.6, 29.1, 25.7, 27.2, 22.9]  # in ps
errors = [0.2, 0.1, 0.3, 0.3, 0.2]

# Field configurations (drift/amplification in kV/cm)
field_configs = ["33.3/22.6", "34.0/22.6", "35.3/22.6", "35.3/21.4", "36.6/21.4"]

# Plot
plt.figure(figsize=(10, 6))
for i, config in enumerate(field_configs):
    plt.errorbar(drift_fields[i], timing_resolutions[i], yerr=errors[i],
                 fmt='o', label=f'Config {config}', capsize=5)

plt.xlabel("Drift Field [kV/cm]")
plt.ylabel("Timing Resolution [ps]")
plt.title("Timing Resolution vs Drift Field")
plt.grid(True)
plt.legend(title="Field Config (Drift/Ampl.)")
plt.tight_layout()
plt.show()
```

```python

# Set ROOT-like style
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["DejaVu Serif"],  # fallback serif
    "font.size": 14,
    "axes.linewidth": 1.2,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 6,
    "ytick.major.size": 6,
    "xtick.minor.size": 3,
    "ytick.minor.size": 3,
    "xtick.top": True,
    "ytick.right": True,
    "legend.frameon": True,
    "legend.edgecolor": "black"
})

# Raw data
drift_fields = [33.3, 34.0, 35.3, 35.3, 36.6]
ampl_fields = [22.6, 22.6, 22.6, 21.4, 21.4]
timing_resolutions = [33.6, 29.1, 25.7, 27.2, 22.9]
errors = [0.2, 0.1, 0.3, 0.3, 0.2]

# Organize data by unique amplification field
unique_ampl_fields = sorted(set(ampl_fields))
data_by_ampl = {a: {"x": [], "y": [], "err": [], "indices": []} for a in unique_ampl_fields}

for idx, (x, a, y, e) in enumerate(zip(drift_fields, ampl_fields, timing_resolutions, errors)):
    data_by_ampl[a]["x"].append(x)
    data_by_ampl[a]["y"].append(y)
    data_by_ampl[a]["err"].append(e)
    data_by_ampl[a]["indices"].append(idx)

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
colors = {22.6: 'darkblue', 21.4: 'darkgreen'}
markers = {22.6: 'o', 21.4: 's'}

# Highlight the highest drift field point in red
max_drift_index = np.argmax(drift_fields)
highlight_x = drift_fields[max_drift_index]
highlight_y = timing_resolutions[max_drift_index]
highlight_err = errors[max_drift_index]

for a_field in unique_ampl_fields:
    d = data_by_ampl[a_field]
    for i, (x, y, e, idx) in enumerate(zip(d["x"], d["y"], d["err"], d["indices"])):
        color = 'red' if idx == max_drift_index else colors[a_field]
        marker = markers[a_field]
        ax.errorbar(x, y, yerr=e, fmt=marker, color=color,
                    capsize=5, markersize=6, label=None if idx != d["indices"][0] else f'Ampl. Field: {a_field} kV/cm')

ax.set_xlabel("Drift Field [kV/cm]")
ax.set_ylabel("Timing Resolution [ps]")
ax.set_title("Timing Resolution vs Drift Field", fontsize=16)
ax.grid(False)
ax.legend(title="Amplification Field", fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
